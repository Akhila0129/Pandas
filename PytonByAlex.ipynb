```python
import pandas as pd
df_flavours=pd.read_csv('Flavors.csv')
df_flavours
```

```python
#Grouping wrt base flavor
abc=df_flavours.groupby('Base Flavor')
```

```python
df_flavours.groupby('Base Flavor').count()
```

```python
df_flavours.groupby('Base Flavor').min()
```

```python
df_flavours.groupby('Base Flavor').max()
```

```python
df_flavours.groupby('Base Flavor').sum()
```

```python
#Obtain all the mean,min,max,count,sum functions at a time for Flavor Rating col
#To do that we need to use agg() functions. We need to pass a dictinary only as a parameter
df_flavours.groupby('Base Flavor').agg({'Flavor Rating':['mean','sum','min','max','count']})
```

```python
#Do the same thing, but this time I need to extract all the aggreagate functions for both flavour ratings and Texture rating
df_flavours.groupby('Base Flavor').agg({'Flavor Rating':['mean','sum','min','max','count'],'Texture Rating':['mean','sum','min','max','count']})
```

```python
#GROUP BY MULTIPLE COLUMNS
df_flavours.groupby(['Base Flavor','Liked']).sum()
#follow same for min(), max(),count()
```

```python
#we can also get all functions at a time for multiple grouping
df_flavours.groupby(['Base Flavor','Liked']).agg({'Flavor Rating':['min','max','sum','count']})
```

```python
df_flavours.groupby('Base Flavor').describe()
```

# **DATA CLEANING USING PANDAS**

```python
import pandas as pd
from google.colab import files
uploaded=files.upload()
```

```python
df=pd.read_excel('Customer Call List.xlsx')
df
```

```python
#clean and standardize l-name, phone num
#Remove duplicate
df=df.drop_duplicates()
df
```

```python
#drop the unnecessary columns
df=df.drop(columns='Not_Useful_Column')
df
```

```python
#cleaning the Last_Name col using strip,lstrip,rstrip and finally in a single step
##f["Last_Name"]=df["Last_Name"].str.strip()
#df["Last_Name"]=df["Last_Name"].str.lstrip("...")
#df["Last_Name"]=df["Last_Name"].str.lstrip("/")
#df["Last_Name"]=df["Last_Name"].str.rstrip("_")
df["Last_Name"]=df["Last_Name"].str.strip("123./_")
df
```

```python
#cleaning the phone number.
#Step-1: extract all the numeric values(i.e., ignore -,/ or anyother)
#step-2: Format or standardise the result of step1
df["Phone_Number"]=df["Phone_Number"].str.replace('[^a-zA-Z0-9]','',regex=True)
df
```

```python
#Format
#df["Phone_Number"].apply(lambda x: x[0:3]+'-'+x[3:6]+'-'+x[6:10])   => throws an error bcz phone num is a float.
                                                                        #We need to convert it to string and then apply
df["Phone_Number"]=df["Phone_Number"].apply(lambda x:str(x))
df["Phone_Number"]=df["Phone_Number"].apply(lambda x: x[0:3]+'-'+x[3:6]+'-'+x[6:10])
df
```

```python
#Step3- get rid of non numeric content likt NaN.. in phone num col
df["Phone_Number"]=df["Phone_Number"].str.replace('nan--','')
df["Phone_Number"]=df["Phone_Number"].str.replace('Na--','')
df
```

```python
#split the address based on ','
df[["Street Address","State","Zipcode"]]=df["Address"].str.split(pat=',',n=2,expand=True)
df
```

```python
#Replace Y with Yes and N with No
df["Paying Customer"]=df["Paying Customer"].str.replace('Yeseseseses','Y')
df["Paying Customer"]=df["Paying Customer"].str.replace('Nooooo','N')
df
```

```python
df["Do_Not_Contact"]=df["Do_Not_Contact"].str.replace('Yeses','Y')
df["Do_Not_Contact"]=df["Do_Not_Contact"].str.replace('Noo','N')
df
```

```python
df["Do_Not_Contact"]=df["Do_Not_Contact"].str.replace('Yes','Y')
df["Do_Not_Contact"]=df["Do_Not_Contact"].str.replace('No','N')
df
```

```python
#clean the nan,na,n/a data
df.fillna('')
```

```python
#Extract the phone numbers to contact
for x in df.index:
  if df.loc[x,"Do_Not_Contact"]=='Y':
    df.drop(x,inplace=True)
df
```

```python
for x in df.index:
  if df.loc[x,"Phone_Number"]=='':
    df.drop(x,inplace=True)
df
```

```python
#df.reset_index()
df=df.reset_index(drop=True)
df
```
